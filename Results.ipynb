This notebook aims to visualize the results obtained from the experiment

# Import Libraries

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the dataset

```python
df = pd.read_csv("data/results2.csv")
```

```python
df=df.iloc[:,1:]
df
```

```python
for k in (10,20,30):
    precision = df[f"MP@{k}"]
    recall = df[f"MR@{k}"]
    df[f"F1@{k}"]=(precision*recall)/(precision+recall)
    
```

```python
df
```

```python
df["model"]=["ViT-Base","SWIN-Tiny","DeiT-Base","Resnet-50"]
df
```

# Visualizations

```python
#set the seaborn style
sns.set_style("whitegrid")

#set the plot size
plt.figure(figsize=(10,6))

#plot
sns.lineplot(data=df,x="model",y="F1@10", marker ="o",color="b", label ="F1@10") # F1@10
sns.lineplot(data=df,x="model",y="F1@20",marker = "s", color ="orange", label="F1@20") # F1@20
sns.lineplot(data=df,x="model",y="F1@30",marker = "^", color ="red", label = "F1@30") #F1@30

plt.xlabel("Models")
plt.ylabel("F1 Score")
plt.title("Models vs F1 Scores")
plt.legend()
```

## Precision

```python
 

#set the seaborn style
sns.set_style("whitegrid")

#set the plot size
plt.figure(figsize=(10,6))

#plot
sns.lineplot(data=df,x="model",y="MP@10", marker ="o",color="b", label ="MP@10") # F1@10
sns.lineplot(data=df,x="model",y="MP@20",marker = "s", color ="orange", label="MP@20") # F1@20
sns.lineplot(data=df,x="model",y="MP@30",marker = "^", color ="red", label = "MP@30") #F1@30

plt.xlabel("Models")
plt.ylabel("Mean Precision Score")
plt.title("Models vs Mean Precision Scores")
plt.legend()
```
